```python
import pandas as pd

df = pd.read_excel("C:/projects/Ecommerce_Segmentation_Project/Data/Online Retail.xlsx")
# Remove canceled invoices
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Remove negative quantities
df = df[df['Quantity'] > 0]

# Remove missing customer IDs
df = df[df['CustomerID'].notna()]

# Remove invalid prices
df = df[df['UnitPrice'] > 0]

# Convert InvoiceDate
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create TotalPrice
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Keep only United Kingdom
df = df[df['Country'] == 'United Kingdom']

# Reset index
df.reset_index(drop=True, inplace=True)
df.to_csv("C:/projects/Ecommerce_Segmentation_Project/Data/cleaned_data.csv", index=False)
```
